```python
# id get only numbers / delete auction (first N characters)
df = df.rename(columns={'id': 'auction_id'}) #change column name 
# df['auction_id'] = df['auction_id'].str.replace("Auction  ",'')
df['auction_id'] = df['auction_id'].str.extract('(\d+)', expand=False)
```

```python
# split date and location by comma
new = df["date"].str.split(",", n = 1, expand = True) 
df["date"] = new[0] 
df["location"] = new[1] 
```

```python
# nooflotsget three first numbers
# df['nooflots'].unique()
# df['nooflots'] = df['nooflots'].str[:3]
df['nooflots'] = df['nooflots'].str.extract('(\d+)', expand=False)
```

```python
df['lot'] = df['lot'].str.replace("Lot ",'')
```

```python
df['currency'] = df['currency'].str.replace('[^a-zA-Z]', '')
```

```python
# total sale get only number / delete sale total% and %(GBP) and " "
df['total_sale'] = df['total_sale'].str.replace("SALE TOTAL: ",'')
new = df["total_sale"].str.split("(", n = 1, expand = True) 
df["total_sale"] = new[0] 
# df['total_sale'] = df['total_sale'].str[:5]
```

```python
# get estimate numbers
# df['estimated'] = (df['estimated'].str.strip('£'))#.astype(int))
# split min and max + max .str[symbol:]
new = df["estimated"].str.split("-", n = 1, expand = True) 
df["min_estimated"] = new[0] 
df['min_estimated'] = df['min_estimated'].str[10:]
df["max_estimated"] = new[1] 
df['max_estimated'] = df['max_estimated'].str[2:]
new = df["max_estimated"].str.split("(", n = 1, expand = True) 
df["max_estimated"] = new[0] 
df.drop('estimated', inplace=True, axis=1)
# get price realized numbers
df['realized'] = df['realized'].str.replace("price realized £",'')
new = df["realized"].str.split("(", n = 1, expand = True) 
df["realized"] = new[0] 
```

```python
df['total_sale'] = df['total_sale'].str.replace(",",'')
df["total_sale"] = pd.to_numeric(df["total_sale"])
```

```python
# i tried with a library but the result didn't work properly so I write todays currency manually
# this is not the optimized way of doing it but for now it solves the problem 
exchange_rate_eur = 1.17
exchange_rate_usd = 1.33
exchange_rate_gbp = 1.00
```

```python
df['total_sale'] = np.where(df['currency']=='USD', 
                           df['total_sale']*exchange_rate_usd, 
                           df['total_sale'])
df['total_sale'] = np.where(df['currency']=='EUR', 
                           df['total_sale']*exchange_rate_eur, 
                           df['total_sale'])
df['realized'] = np.where(df['currency']=='USD', 
                           df['realized']*exchange_rate_usd, 
                           df['realized'])
df['realized'] = np.where(df['currency']=='EUR', 
                           df['realized']*exchange_rate_eur, 
                           df['realized'])
df['min_estimated'] = np.where(df['currency']=='USD', 
                           df['min_estimated']*exchange_rate_usd, 
                           df['min_estimated'])
df['min_estimated'] = np.where(df['currency']=='EUR', 
                           df['min_estimated']*exchange_rate_eur, 
                           df['min_estimated'])
df['max_estimated'] = np.where(df['currency']=='USD', 
                           df['max_estimated']*exchange_rate_usd, 
                           df['max_estimated'])
df['max_estimated'] = np.where(df['currency']=='EUR', 
                           df['max_estimated']*exchange_rate_eur, 
                           df['max_estimated'])
```
